```python
mask = "<extra_id_X>"
for i in range(len(data)):
    row = data[i].strip()
    splitted_row = row.splitlines()
    splitted_row = [el.split() for el in splitted_row]
    tokens = [el[0] for el in splitted_row]
    pos = [pos_map[el[1]] for el in splitted_row]
    sentence = ' '.join(tokens)
    out = []
    cnt = 0
    for j in range(len(tokens)):
        cnt = cnt%100
        tup = f"{mask.replace('X',str(cnt))} {tokens[j]} {mask.replace('X',str(cnt+1))} {pos[j].lower()}"
        cnt += 2
        out.append(tup)
    out = " ; ".join(out)
    inputs = f"Extract POS TAG with format >> pos : <extra_id_0>, tag : <extra_id_1> | {sentence}"

    data[i] = {
        "input" : inputs,
        "output" : out
    }
```
